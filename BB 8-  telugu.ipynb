## Problem Statement
      Bigg Boss is a reality television series where contestants compete for the title by surviving weekly eliminations based on their behavior, tasks, and audience votes. Understanding the factors influencing a contestant's success is challenging due to the interplay of personal attributes, audience perception, and in-show dynamics. The primary goal of this project is to predict whether a contestant will win the show based on available historical data. 

## Features:
1. Language                         : The language in which the Bigg Boss season was conducted (e.g., Hindi, Telugu, Tamil, etc.).(Categorical)
2. Season Number                    : The season number of the show within a particular language (e.g., 1 for the first season).(Numerical)
3. Name                             : The name of the contestant who participated in the season.(Text)
4. Profession                       : The occupation of the contestant (e.g., Actor, Social Media Star, Singer).(Categorical/Text)
5. Gender                           : The gender of the contestant (e.g., Male, Female, Others).(Categorical)
6. Age                              : The age of the contestant at the time of their participation.(Numerical)
7. Entry Date                       : The date on which the contestant entered the Bigg Boss house.(Date)
8. Elimination Date                 : The date on which the contestant was eliminated from the show.(Date)
9. Elimination Week Number          : The week number during the season when the contestant was eliminated.(Numerical)
10. Contestant Number               : A unique identifier or ranking assigned to a contestant within a season.(Numerical)
11. Wild Card                       : Indicates whether the contestant entered the house as a wild-card entry (Yes/No).(Categorical)
12. Season Length                   : The total duration of the season (e.g., in days).(Numerical)
13. Number of Housemates            : The total number of contestants in that season.(Numerical)
14. Season Start Date               : The date on which the season started.(Date)
15. Season End Date                 : The date on which the season ended.(Date)
16. Host Name                       : The name of the host who presented that season of Bigg Boss.(Text)
17. Guest Host Name                 : The name of any guest host(s) who temporarily hosted the season.(Text)
18. Prize Money (INR)               : The prize money awarded to the winner of the season (in Indian Rupees).(Numerical)
19. Broadcasted By                  : The TV network/channel on which the season was broadcasted (e.g., Colors, Star Maa).(Text)
20. OTT Platform                    : The online platform (if any) where the show was streamed (e.g., Voot, Disney+ Hotstar).(Text)
21. Average TRP                     : The average Television Rating Points (TRP) for the season, which measures viewership.(Numerical)
22. Most Viewed States              : The states in India where the season had the highest viewership.(Text/List)
23. House Location                  : The physical location of the Bigg Boss house for that season.(Text)
24. OTT Season                      : Indicates whether the season was primarily an OTT (Over-the-Top) release (Yes/No).(Categorical)
25. Season Slogan                   : The tagline or slogan used to promote the season.(Text)
26. Season Theme                    : The central theme or concept of the season (e.g., Jungle, VVIP, Double Trouble).(Text)
27. Number of Evictions Faced       : The total number of eviction nominations faced by the contestant.(Numerical)
28. Number of Times Elected as Captain : The number of times a contestant was elected as the house captain.(Numerical)
29. Number of Re-entries            : The number of times a contestant re-entered the house after being eliminated.(Numerical)
30. Walked                          : Indicates whether the contestant voluntarily walked out of the show (Yes/No).(Categorical)
31. Ejected                         : Indicates whether the contestant was forcibly ejected from the show (Yes/No).(Categorical)
32. In Secret-Room                  : Indicates whether the contestant spent time in the secret room (Yes/No).(Categorical)
33. Won Eviction Free Pass          : Indicates whether the contestant won an eviction-free pass (Yes/No).(Categorical)
34. Won Ticket to Finale            : Indicates whether the contestant won the "Ticket to Finale" (Yes/No).(Categorical)
35. Number of Times Best Performer  : The number of times the contestant was declared the best performer.(Numerical)
36. Number of Times Worst Performer : The number of times the contestant was declared the worst performer.(Numerical)
37. Secret Task                     : Indicates whether the contestant successfully completed a secret task (Yes/No).(Categorical)
38. Social Media Popularity         : A metric or score reflecting the contestant’s popularity on social media platforms.(Numerical)
39. Finalist                        : Indicates whether the contestant reached the finals of the season (Yes/No).(Categorical)
40. Winner                          : Indicates whether the contestant won the season (Yes/No).(Categorical)

## 🔍 Preparing Data for ML modelling

```python
# Handling NULL values
bigg_boss.isnull().sum()
```

```python
# Removing records where Name field is empty
bigg_boss = bigg_boss.loc[bigg_boss.Name.notnull()]
bigg_boss.reset_index(drop=True,inplace=True)
```

```python
# Contestant might have faced at least one eviction, so filling NaN with 'Number of Evictions Faced' with 1
bigg_boss['Number of Evictions Faced'] = bigg_boss['Number of Evictions Faced'].fillna(1)

# Number of re-entries are very less, so filling NULLs in 'Number of re-entries' with 0
bigg_boss['Number of Re-entries'] = bigg_boss['Number of Re-entries'].fillna(0)

# Filling blank values in 'Average TRP' column with average
bigg_boss['Average TRP'] = bigg_boss['Average TRP'].fillna(bigg_boss['Average TRP'].mean())

# Filling NULLs in 'Number of times elected as Captain' with 0
bigg_boss['Number of Times Elected as Captain'] = bigg_boss['Number of Times Elected as Captain'].fillna(0)
```

```python
bigg_boss['Season Starting Year'] = bigg_boss['Season Start Date'].dt.year.astype(pd.Int32Dtype())
bigg_boss['Season Ending Year'] = bigg_boss['Season End Date'].dt.year.astype(pd.Int32Dtype())
bigg_boss['Season Starting Quarter'] = bigg_boss['Season Start Date'].dt.quarter.astype(pd.Int32Dtype())
bigg_boss['Season Ending Quarter'] = bigg_boss['Season End Date'].dt.quarter.astype(pd.Int32Dtype())
bigg_boss.info()
```
