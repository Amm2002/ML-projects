```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
from sklearn.ensemble import RandomForestClassifier
```

```python
X = Avi_cleaned.drop(['Total.Fatal.Injuries'], axis=1)
y = Avi_cleaned['Total.Fatal.Injuries'] > 0
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
model= RandomForestClassifier()
model
```

```python
Avi=Avi.apply(pd.to_numeric,errors='coerce')
```

```python
Columns_convert = ['Total.Fatal.Injuries']
Avi[Columns_convert]=Avi[Columns_convert].apply(pd.to_numeric,errors='coerce')
```

```python
Avi.columns
```

```python
Avi_cleaned = Avi.drop(['Event.Id', 'Accident.Number', 'Event.Date', 'Location', 'Country', 'Airport.Code', 'Airport.Name', 'Registration.Number', 'Make', 'Model', 'Air.carrier', 'Publication.Date'], axis=1)
```

```python
Avi_cleaned
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print('Classification Report:',classification_report(y_test,y_pred))
```
